## 아파트 실거래가 분석

```python
import pandas as pd

apt_price = pd.read_csv(
    './apart_price.csv',
    encoding = 'cp949',
    skiprows = 15,
    thousands=',',
#     dtype = ['건축년도':'int32']
)
apt_price
```

```python
apt_price.info()
```

```python
apt_price.shape
```

```python
apt_price['시군구'].str.split()
```

```python
apt_price['구'] = apt_price['시군구'].str.split().str[1]
apt_price['동'] = apt_price['시군구'].str.split().str[2]

apt_price.head()
```

```python
# 전용면적을 평으로 변환
apt_price['평'] = apt_price['전용면적(㎡)'] / 3.03

apt_price.head()
```

```python
# 거래금액 단위를 억원으로 변환

apt_price['거래금액_억원'] = apt_price['거래금액(만원)'] / 10000
apt_price.head()
```

```python
# 계약 년, 월 분리

apt_price['계약년'] = apt_price['계약년월'] // 100
apt_price['계약월'] = apt_price['계약년월'] % 100

apt_price.head()
```

```python
# 평정보를 이용하여 10평이하, 10평대, 20평대, 30평대, 40평대, 50평대 이상의 평형정보 생성

apt_price['평형'] = pd.cut(
    apt_price['평'],
    bins = [0, 10, 20, 30, 40, 50, 1000],
    labels = '10평이하, 10평대, 20평대, 30평대, 40평대, 50평대 이상'.split(', ')
)
apt_price['평형']
```

```python
# 요일 추출

pd.to_datetime(
    apt_price['계약년월'].astype(str) + apt_price['계약일'].astype(str),
    format = '%Y%m%d'
)
```

```python
apt_price['계약요일'] = pd.to_datetime(
    apt_price['계약년월'].astype(str) + apt_price['계약일'].astype(str),
    format = '%Y%m%d'
).dt.dayofweek



apt_price.head()
```

```python
# 평정보를 평형으로 전환 (apply 활용)

apt_price['평'].describe()
def 평형변환(x):
    if x < 10:
        return '10평이하'
    if x < 20:
        return '10평대'
    if x < 30:
        return '20평대'
    if x < 40:
        return '30평대'
    if x < 50:
        return '40평대'
    if x > 50:
        return '40평이상'
    
apt_price['평'].apply(평형변환)
```
