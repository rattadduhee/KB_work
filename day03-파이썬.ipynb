```python
def mySub(num1, num2):
    if(num1 > num2):
        return num1 - num2
    else:
        return num2 - num1

mySub(1,3)
```

```python
def mySum(end, start=0):
    sum = 0
    for i in range(start, end+1):
        sum += 1
    return sum

def mySum1(*args):     # 매개변수 앞에 *가 붙으면 list형태로 전달
    sum = 0
    for i in args:
        sum += i
    return sum

print(mySum(10), mySum(10,5))
print(mySum(start=5,end=10))

print(mySum1(1,2,3,4))
s = [1,2,3,4]
print(mySum1(*s))
```

```python
def printInfo(**args):    # 매개변수 앞에 **가 붙으면 dict형태로 전달
    print(args)
    
printInfo(a=1, b=2)

dict = {'a':3, 'b':4}
printInfo(**dict)
```

```python
def my_plus(a, *args):
    print(a)
    print(args)
    
my_plus(10,20,30)     # 첫번째 변수는 a로, 나머지는 args로 전달됨
```

```python
# 구구단 출력 함수

def gugudan(num=2, num2=0):
    """구구단 출력함수"""
    if num2 == 0:
        num2 = num
        
    for j in range(num, num2+1):
        for i in range(1,10):
            print(j * i, end=" ")
    #         print(f'{num} * {i} = {num*i}')
        print()
        
gugudan(3)
gugudan(4)
gugudan()        # 2단 출력
gugudan(3,5)     # 3단~5단 출력
```

```python
# 회원등록 함수

def add_member(*args):
    s = []
    print('멤버수는',len(args))
    s.extend(args)
#     s = args
    return s
    
print(add_member('홍길동'))
print(add_member('홍길동', '김길동'))
print(add_member('홍길동', '김길동', '강길동'))
```

```python
# 회원등록 함수 (강사님 풀이) - 클로저(Closure)

def add_member():
    members = []
    count = 0
    
    def add(*mems):
        nonlocal count    # global과 local사이의 변수
        members.extend(mems)
        count = count + len(mems)  
    
    def print_info():
        print("count =",count)
        
    return add, print_info
    
a, p = add_member()     # print_info() 함수의 주소
a('홍길동')
a('홍길동', '김길동')
p()

# p = add_member('홍길동', '김길동', '강길동')
# p()
```

# 지뢰찾기 게임

```python
# 지뢰찾기 (빨리찾아 터트리기)
# 필드 (2차원 배열)
import random

fields = list()
res = list()
count = 10
try_count = 0

def print_arr(a):     # display() 함수 활용가능
    for i in res:
        for j in i:
            print(j, end=' ')
        print()

for i in range(10):
    fields.append([0]*10)
    res.append(['-']*10)

nums = list(range(100))
for i in random.sample(nums,10):
    x = i // 10
    y = i % 10
    fields[x][y] = 9  # 지뢰위치
        
    # 8방위치 1증가
    if x < len(fields) - 1 :
        fields[x+1][y] += 1
        if y < len(fields) - 1 :
            fields[x+1][y+1] += 1  
        if y > 0 :
            fields[x+1][y-1] += 1
    if y > 0 :
        fields[x][y-1] += 1
    if y < len(fields) - 1 :
        fields[x][y+1] += 1
    if x > 0 :
        fields[x-1][y] += 1
        if y < len(fields) - 1 :
            fields[x-1][y+1] += 1
        if y > 0 :
            fields[x-1][y-1] += 1
            
print_arr(res)
    
while count > 0 :
    print()
    str = (input('좌표를 입력하세요(x,y): ')).split(',')
    row = int(str[0]) - 1
    col = int(str[1]) - 1
    try_count += 1
    if(fields[row][col] >= 9):
        res[row][col] = 'X'
        print_arr(res)
        count -= 1
        print(f'지뢰발견! {count}개의 지뢰가 남았어요!')        
    else:        
        res[row][col] = fields[row][col]
        print_arr(res)

print(f'{try_count}번만에 지뢰를 발견했어요!')
```
