## 아파트 실거래가 데이터 찾기

```python
import pandas as pd

apt_price = pd.read_csv(
    './apt_price_2022.csv', # 파일명
    encoding = 'cp949',     # 한글 인코딩
    skiprows = 15,          # 15줄 설명부분 삭제
    thousands = ','         # 숫자사이 ',' 삭제
)

apt_price
```

```python
# str : string의 accessor
apt_price['시군구'].str.split()  
```

```python
# '시군구'에서 구를 분리하여 '구'컬럼에 저장

apt_price['구'] = apt_price['시군구'].str.split().str[1]
apt_price['구']
```

```python
apt_price['전용면적(㎡)']
```

```python
# 구별 추출, 통계
# 강남구 거래건만 추출

apt_price[apt_price['구']=='강남구'].shape    # (1054, 16)
apt_price[apt_price['구']=='강남구']

# 강남 3구(강남구, 송파구, 서초구)의 거래건수는 몇건 ?
apt_price[apt_price['구'].isin(['강남구', '송파구', '서초구'])].shape
```

```python
apt_price['동'] = apt_price['시군구'].str.split().str[2]
apt_price['동']
```

```python
# 거래건수가 제일 많은 구 이름과 건수를 출력하세요
max_name = ''
max_count = 0

# 거래건수저장용 배열 생성
count_arr = []

# 사용된 구이름 목록 추출(구컬럼.unique)
구이름 = apt_price['구'].unique()

# 구 이름별 행추출, 건수 배열에 누적
for 구 in 구이름 :
    거래건수 = apt_price[apt_price['구'] == 구].shape[0]
    if 거래건수 > max_count:
        max_name = 구
        max_count = 거래건수

print(f'거래건수가 가장 많은 구 : {max_name}, {max_count}번')
```

```python
계약날짜 = apt_price['계약년월'].astype(str) + apt_price['계약일'].astype(str)
apt_price['계약요일'] = pd.to_datetime(계약날짜, format = '%Y%m%d').dt.dayofweek
apt_price['계약요일']
```

```python
# 평방미터를 평으로 변환
apt_price['평'] = apt_price['전용면적(㎡)'] / 3.03
apt_price['평']
```

```python
# 평을 크기구간으로 나누기
# 10평미만, 10평대, 20평대, 30평대, 40평대, 50평이상
apt_price['평형'] = pd.cut(
    apt_price['평'],
    bins = [0, 10, 20, 30, 40, 50, 100],
    labels = ['10평미만', '10평대', '20평대', '30평대', '40평대', '50평이상']
)
apt_price.head()
```

```python
apt_price.groupby('구').mean()
```

```python
apt_price['거래금액(만원)'].describecribe()
```

```python
#거래금액(만원) 의 값을 기준으로
# 1억미만, 5억미만, 10억미만, 20억미만, 50억미만, 50억이상 의 
# 구간으로 나눠보세요. "가격대" 칼럼에 저장

apt_price['가격대'] = pd.cut( 
    apt_price['거래금액(만원)'],
    bins = [0, 10000, 50000, 100000, 200000, 500000, 10000000],
    labels = '1억미만,5억미만,10억미만,20억미만,50억미만,50억이상'.split(',')
)

apt_price
```

```python
apt_price.groupby('가격대')['거래금액(만원)'].count()
```

```python
# 강남3구의 평형별 평균 거래가와
# 강남3구 이외 지역의 평형별 평균 거래가
apt_price[apt_price['구'].isin(['강남구','송파구','서초구'])].groupby('평형')['거래금액(만원)'].mean()
apt_price[~apt_price['구'].isin(['강남구','송파구','서초구'])].groupby('평형')['거래금액(만원)'].mean()
```

```python
# 구별, 평형별 평균거래금액
apt_price.groupby(['구','평형'])[['거래금액(만원)']].mean()
```

```python
# 구별, 평형별 평균거래금액

apt_price.pivot_table(
    index="구",
    columns="평형",
    values="거래금액(만원)"
)
```

```python
apt_price.pivot_table(
    index="구",
    columns="평형",
    values="거래금액(만원)",
    aggfunc=["min","max"]
)
```
