### 웹페이지
    1. 정적페이지 : 소스보기와 페이지가 같은 내용을 가짐
                    BeautifulSoup으로 추출 가능
    2. 동적페이지 : 소스보기와 페이지가 다른 내용을 가짐
                    자바 스크립트가 태그를 생성
                    BeautifulSoup으로 추출 불가능 → Selenium 사용

```python
import requests
from bs4 import BeautifulSoup

# 환율 크롤링
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%99%98%EC%9C%A8'
res = requests.get(url).text

bs = BeautifulSoup(res, 'html.parser')

tbl_list = bs.select('table.rate_table_info > tbody > tr')

res_list = {}

for i in tbl_list:
    th = i.select_one('th').get_text()
    td = i.select_one('td').get_text()
    res_list[th] = td

res_list
```

### 문자열 패턴 -> Regular Expression : 정규식

[중요 패턴 기호]
* . : 1개의 아무글자
* ? : 왼쪽글자의 0번또는 1번 반복
* /* : 왼쪽글자의 0번 또는 여러번 반복

```python
# 출력 예 : [{'일시' : '07.01(금)', '시간':'18:30'}]
# ['rows'][num]['row'][num][text]
import re
lis = sched_json['rows']

data_lis = {}
    
for i in range(len(lis)):
    day_lis = []
    for j in range(len(lis[i]['row'])):
        s = re.sub('<.*?>', '', lis[i]['row'][j]['Text'], 0).strip()
        day_lis.append(s)
    if len(day_lis) > 8 :
        temp = day_lis[0]
        data_lis['일시'] = temp
        data_lis['시간'] = day_lis[1]
        score = ':'.join(re.findall(r'\d+',day_lis[2]))
        data_lis['팀'] = re.sub('vs', ' vs ',day_lis[2]) if score == '' else re.sub('\d+vs\d+', ' vs ',day_lis[2])
        data_lis['점수'] = '미실시' if score == '' else score 
        data_lis['장소'] = day_lis[7]
    else :
        data_lis['일시'] = temp
        data_lis['시간'] = day_lis[0]
        score = ':'.join(re.findall(r'\d+',day_lis[1]))
        data_lis['팀'] = re.sub('vs', ' vs ',day_lis[1]) if score == '' else re.sub('\d+vs\d+', ' vs ',day_lis[1])
        data_lis['점수'] = '미실시' if score == '' else score 
        data_lis['장소'] = day_lis[6]
    print(data_lis)
```
