# 학생점수와 출석일수

```python
import pandas as pd

data = pd.read_excel('student_info.xls')
data
```

```python
# 데이터 복사본
data2 = data.copy()

# 거리계산을 위해 정규화 전처리 진행
from sklearn.preprocessing import MinMaxScaler
data2[['Mark','Attended']] = MinMaxScaler().fit_transform(
    data2[['Mark','Attended']]
)
data2
```

```python
import matplotlib.pyplot as plt

# 군집분석
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2)
pred = model.fit_predict(data2[['Mark','Attended']])

plt.scatter( 
    data["Mark"], 
    data["Attended"],
    c = pred # color : 같은 값끼리 같은 색을 가짐
) 
```

```python
import matplotlib.pyplot as plt

# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,6))
for k in range(2,8):
    model = KMeans(n_clusters = k)
    pred = model.fit_predict(data2[['Mark','Attended']])
    plt.subplot(2,3,k-1)
    plt.title(f'K={k}')
    plt.xlabel('Mark')
    plt.ylabel('Attended')
    plt.scatter( 
        data["Mark"], 
        data["Attended"],
        c = pred # color : 같은 값끼리 같은 색을 가짐
    ) 
```

```python
# 군집분석
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
pred = model.fit_predict(data2[['Mark','Attended']])

#군집별 색깔 지정
plt.scatter( 
    data["Mark"], 
    data["Attended"],
    c = pred                # color : 같은 값끼리 같은 색을 가짐
) 

#각 데이터에 제품번호 출력
for index,semester, name, course, mark, attend in data.itertuples():
    plt.annotate( f"{pred[index]}:{name}",(mark+0.5, attend+0.5))
plt.show()
```

```python
# 학생별 점수와 출석의 평균
data_grouped = data.groupby('Name')[['Mark','Attended']].mean()
data_grouped
```

```python
# 거리계산을 위해 정규화 전처리 진행
from sklearn.preprocessing import MinMaxScaler
data_grouped[['Mark','Attended']] = MinMaxScaler().fit_transform(
    data_grouped[['Mark','Attended']]
)
data_grouped
```

```python
import matplotlib.pyplot as plt

# 군집분석
from sklearn.cluster import KMeans
# 여러가지 군집 갯수를 테스트
plt.figure(figsize=(10,6))
for k in range(2,8):
    model = KMeans(n_clusters = k)
    pred = model.fit_predict(data_grouped[['Attended', 'Mark']])
    plt.subplot(2,3,k-1)
    plt.title(f'K={k}')
    plt.xlabel('Attended')
    plt.ylabel('Mark')
    plt.scatter( 
        data_grouped["Attended"], 
        data_grouped["Mark"],
        c = pred # color : 같은 값끼리 같은 색을 가짐
    ) 
```

```python
# 군집분석
model = KMeans(n_clusters=3)
pred = model.fit_predict(data_grouped[["Attended","Mark"]])
# 2행 3열 그래프
plt.figure(figsize=(10,6)) #그래프 크기 지정
plt.subplot(2,3,k-1) #그래프 순번은 1부터 시작
plt.tight_layout()
plt.title(f"K={k}")
plt.xlabel("Attended")
plt.ylabel("Mark")
plt.scatter( 
    data_grouped["Attended"], 
    data_grouped["Mark"],
    c = pred # color : 같은 값끼리 같은 색을 가짐
) 
#각 데이터에 제품번호,지역번호 출력
for name, mark, attended in data_grouped.itertuples():
    plt.annotate( f"{name}",(attended+0.05,mark)  )
plt.show()
```
